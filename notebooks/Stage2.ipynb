# Analyzing Data

```python
#Its is a input
top = 1000
#Considering all position values over 'top' value, grouping Worth by Industry, and sorting values 
ind_groups = proc_data[ proc_data['position'] <= top ].groupby('Industry')['worth_BUSD'].sum().sort_values(ascending=False)
#Converting 'Industry' as index to column, to represent in bar plot 
ind_groups = ind_groups.reset_index()
```
